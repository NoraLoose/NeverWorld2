```python
import matplotlib.pyplot as plt
import numpy
%matplotlib inline
```

```python
# Density profile used in Neverworld (Khani et al., 2019)
h = numpy.array([150, 250, 600, 1000, 1000, 1000]) # Initial layer thickness
r = numpy.array([1025.5, 1027.0, 1027.5, 1027.8, 1028.0, 1028.1]) # Layer density
print('Total depth =', h.sum(), '# of layers =', h.shape[0] )

# Interface depths
e = numpy.zeros( h.shape[0]+1 )
e[1:] = h.cumsum()
# Layer centers
ec = ( e[:-1] + e[1:] )/2

plt.figure(figsize=(12,5))
ax = plt.subplot(121)
ax.plot( numpy.vstack([r,r]), -numpy.vstack([e[:-1], e[1:]]), 'k' );
plt.plot(r, -ec, 'ko--')
plt.xlabel('Layer density (kg m$^{-3}$)')
plt.ylabel('z (m)')
plt.ylim(-4000,0)

drhodz = -( r[:-1]-r[1:] )/( ec[:-1]-ec[1:] )
ax = plt.subplot(122)
ax.plot(numpy.sqrt(-0.01*drhodz)/1e-4, -e[1:-1], 'x-')
plt.axvline(numpy.sqrt(0.01*(r[-1]-r[0])/4000)/1e-4, linestyle='--')
plt.xlabel('N/f')
plt.ylim(-4000,0);
```

```python
# New profile for Neverworld2
h2 = numpy.array([25, 50, 100, 125, 150, 175, 200, 225, 250, 300, 350, 400, 500, 550, 600])
print('Total depth =', h2.sum(), '# of layers =', h2.shape[0] )

print('h = ',end='')
for i in range(len(h2)): print(h2[i], end=', ')
print()
e2 = numpy.zeros( h2.shape[0]+1 )
e2[1:] = h2.cumsum()
ec2 = ( e2[:-1] + e2[1:] )/2
def rrr(e, d=1000, drho=5.5):
    fe = numpy.exp((4000-e)/d)
    fe = fe - fe[-1]
    fe = fe/fe[0]
    return 1028.1 - drho * fe
r2 = rrr(ec2).round(2)
print('Layer = ',end='')
for i in range(len(r2)): print(r2[i], end=', ')
print()

plt.figure(figsize=(12,5))
ax = plt.subplot(121)
ax.plot(r2, -ec2, 'o-')
plt.xlabel('Layer density (kg m$^{-3}$)')
plt.ylabel('z (m)')
plt.ylim(-4000,0)

drhodz2 = -( r2[:-1]-r2[1:] )/( ec2[:-1]-ec2[1:] )
ax = plt.subplot(122)
ax.plot(numpy.sqrt(-0.01*drhodz2)/1e-4, -e2[1:-1], 'x-')
plt.axvline(numpy.sqrt(0.01*(r2[-1]-r2[0])/4000)/1e-4, linestyle='--')
plt.xlabel('N/f')
plt.ylim(-4000,0);
```
